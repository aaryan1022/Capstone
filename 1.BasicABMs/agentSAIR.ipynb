```python
import random
import pandas as pd
N = 10000
S = 8000
A = 1000
I = 1000
R = 0
trA = 0.3
trI = 0.6
rr = 0.1
Sc = []
Ic = []
Rc = []
Ac = []

pop = ['S']*S + ['I']*I + ['R']*R + ['A']*A

for t in range(100):
    npop = []
    for ag in pop:

        p = random.random()    
        
        if(ag == 'S'):
            
            for i in range(5):
                p2 = random.randint(0,N-1)
                if(pop[p2]=='I'):
                    if(p<=trI):
                        ag = 'I'
                        break
                elif(pop[p2]=='A'):
                    if(p<=trA):
                        ag = 'A'
                        break

        elif(ag == 'I' or ag == 'A'):

            if(p <= rr):
                ag = 'R'


        else:
            pass

        npop.append(ag)
        
    pop = npop
    S = pop.count('S')
    Sc.append(S)
    I = pop.count('I')
    Ic.append(I)
    R = pop.count('R')
    Rc.append(R)
    A = pop.count('A')
    Ac.append(A)

from matplotlib import pyplot as plt

plt.plot(Sc, label='S')
plt.plot(Ic, label='I')
plt.plot(Rc, label='R')
plt.plot(Ac, label='A')
plt.legend()
plt.show()
```
