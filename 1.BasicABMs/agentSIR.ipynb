```python
import random
import pandas as pd
N = 10000
S = 9999
I = 1
R = 0
tr = 0.5
rr = 0.1
Sc = []
Ic = []
Rc = []

pop = ['S']*S + ['I']*I + ['R']*R

for t in range(100):
    npop = []
    for ag in pop:

        p = random.random()    
        
        if(ag == 'S'):
            
            for i in range(5):
                p2 = random.randint(0,N-1)
                if(pop[p2]=='I'):
                    if(p<=tr):
                        ag = 'I'
                        break

        elif(ag == 'I'):

            if(p <= rr):
                ag = 'R'


        else:
            pass

        npop.append(ag)
        
    pop = npop
    S = pop.count('S')
    Sc.append(S)
    I = pop.count('I')
    Ic.append(I)
    R = pop.count('R')
    Rc.append(R)

from matplotlib import pyplot as plt

plt.plot(Sc, label='S')
plt.plot(Ic, label='I')
plt.plot(Rc, label='R')
plt.legend()
plt.show()
```
