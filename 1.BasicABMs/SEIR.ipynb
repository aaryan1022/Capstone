```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
S0 = 0.95
E0 = 0.1
I0 = 0.05
R0 = 0
a = 0.5
b = 0.2
c = 0.1
Z0 = [S0, E0, I0, R0]
```

```python
def model(t, Z):
    [S, E, I, R] = Z
    dSdt = - a*S*I
    dEdt = a*S*I - b*E
    dIdt = b*E - c*I 
    dRdt = c*I
    return [dSdt, dEdt, dIdt, dRdt]
```

```python
t_span = (0, 100)
t_eval = np.linspace(*t_span, 100)
```

```python
model = solve_ivp(model, t_span, Z0, t_eval=t_eval)
```

```python
plt.plot(model.t, model.y[0], label='S(t)')
plt.plot(model.t, model.y[1], label='I(t)')
plt.plot(model.t, model.y[2], label='R(t)')
plt.plot(model.t, model.y[3], label='E(t)')
plt.legend()
plt.show()
```
