```python
import random
import pandas as pd
```

```python
#add a network strucutre 
#find out where people are
```

```python
N = 1000
S = 800
E = 100
I = 100
R = 0
tr = 0.5
inc = 0.2
rr = 0.1
Sc = []
Ec = []
Ic = []
Rc = []
```

```python
pop = ['S']*S + ['I']*I + ['R']*R + ['E']*E
```

```python
for t in range(100):
    npop = []
    for ag in pop:

        p = random.random()    
        
        if(ag == 'S'):
            
            for i in range(5):
                p2 = random.randint(0,999)
                if(pop[p2]=='I'):
                    if(p<=tr):
                        ag = 'E'
                        break

        elif(ag == 'E'):
            
            if(p <= inc):
                ag = 'I'

        elif(ag == 'I'):

            if(p <= rr):
                ag = 'R'


        else:
            pass

        npop.append(ag)
        
    pop = npop
    S = pop.count('S')
    Sc.append(S)
    E = pop.count('E')
    Ec.append(E)
    I = pop.count('I')
    Ic.append(I)
    R = pop.count('R')
    Rc.append(R)
```

```python
from matplotlib import pyplot as plt

plt.plot(Sc, label='S')
plt.plot(Ec, label='E')
plt.plot(Ic, label='I')
plt.plot(Rc, label='R')
plt.legend()
plt.show()
```
